```python
# Séparation des données en ensembles de formation et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=1) 
X_test
```

```python
X_train
```

```python
# Création du modèle d'arbre de décision
clf = DecisionTreeClassifier(max_depth = 3)
```

```python
# entrainement du modèle
clf.fit(X_train, y_train)

# Prédiction sur l'ensemble de test
y_pred = clf.predict(X_test)
```

```python
X_test
```

```python
cm = confusion_matrix(y_test, y_pred)

# Representation du modèle
# plt.figure(figsize=(8,6))
# sns.heatmap(cm, annot=True, fmt='g',cmap='Blues')
# plt.xlabel('Prediction')
# plt.ylabel('Vraies valeurs')
# plt.title('Matrice de confusion pour la prédiction du cancer du sein')
# plt.show()

df_cm = pd.DataFrame(cm, index=['Cancer du sein non détecté', 'Cancer du sein détecté'], columns=['Non cancer du sein', 'Cancer du sein'])
plt.figure(figsize=(5,4))
sns.heatmap(df_cm, annot=True, cmap='Blues', fmt='g')
plt.xlabel('Valeurs prédites')
plt.ylabel('Valeurs réelles')
plt.title('Tableau de confusion')
plt.show()
```

```python
# Créer un DataFrame pour afficher le résultat
results = pd.DataFrame(columns=['Ligne', 'Prédiction'])

for i in range(len(y_test)):
    results = pd.concat([results, pd.DataFrame({'Ligne': [i], 'Prédiction': [y_pred[i]]})])
    results['Prédiction'] =  results['Prédiction'].replace({0: 'Benigne', 1: 'Maligne'})
    
# Afficher le résultat
print(results)
```

```python
# Calcul de l'exactitude du modèle
accuracy_score(y_test, y_pred)
```

```python
# Evaluation de la capacité du modèle
recall_score(y_test, y_pred)
```

```python
# Mesure de la f1_score du modèle
f1_score(y_test, y_pred)
```

```python
# Mesure de la performance du modèle
roc_auc_score(y_test, y_pred)
```
